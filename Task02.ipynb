```python
# Survival Analysis based on Pclass and Sex:
survival_rate_by_class = train.groupby('Pclass')['Survived'].mean()
survival_rate_by_sex = train.groupby('Sex')['Survived'].mean()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values, color='salmon')
plt.title('Survival Rate by Pclass')

plt.subplot(1, 2, 2)
sns.barplot(x=survival_rate_by_sex.index, y=survival_rate_by_sex.values, color='yellowgreen')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
l = ['C = Cherbourg', 'Q = Queenstown', 'S = Southampton']
s = [0.553571,0.389610,0.336957]
ax.pie(s, labels = l,autopct='%1.2f%%')
plt.title('Percentage of passengers by Embarked countries')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
sns.histplot(train['Age'].dropna(), kde=False)
plt.title('Age Distribution')
plt.subplot(1, 3, 2)
sns.countplot(train['Sex'])
plt.title('Gender Distribution')
plt.subplot(1, 3, 3)
sns.histplot(train['Pclass'].dropna(),kde=True)
plt.title('Pclass distribution')
plt.tight_layout()
plt.show()
```

```python
survived = train['Survived'].value_counts()
alive = survived[1] 
dead = survived[0] 
fig = plt.figure(figsize=(10,4))
ax = fig.add_axes([0,0,1,1])
status = ['Survived','Dead']
ind = [alive,dead]
ax.bar(status,ind, color='teal')
plt.xlabel("Status")
plt.ylabel("No. of passengers")
plt.title("Status of passengers - Survived or Dead")
plt.show()
```
